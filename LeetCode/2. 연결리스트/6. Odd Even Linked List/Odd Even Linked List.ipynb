## 반복 구조로 홀짝 노드 처리

```python
def oddEvenList(self, head):
    # 예외 처리
    if head is None:
        return None
    
    odd = head
    even = head.next
    even_head = head.next
    
    # 반복하면서 홀짝 노드 처리
    while even and even.next:
        odd.next, even.next = odd.next.next, even.next.next
        odd, even = odd.next, even.next
        
    # 홀수 노드의 마지막을 짝수 헤드로 연결
    odd.next = even.head
    return head
```
