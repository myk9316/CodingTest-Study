## 오름차순 풀이

```python
def arrayPairSum(nums):
    pair = []
    sum = 0
    nums.sort()
    
    for n in nums:
        pair.append(n)
        if len(pair) == 2:
            sum += min(pair)
            pair = []
            
    return sum
```

```python
print(arrayPairSum([1,4,3,2]))
print(arrayPairSum([6,2,6,5,1,2]))
```

## 짝수 번째 값 계산

```python
def arrayPairSum(nums):
    sum = 0
    nums.sort()
    
    for i, n in enumerate(nums):
        if i % 2 == 0:
            sum += n
    
    return sum
```

```python
print(arrayPairSum([1,4,3,2]))
print(arrayPairSum([6,2,6,5,1,2]))
```

## 파이썬 다운 방식

```python
def arrayPairSum(nums):
    return sum(sorted(nums)[::2])
```

```python
print(arrayPairSum([1,4,3,2]))
print(arrayPairSum([6,2,6,5,1,2]))
```
