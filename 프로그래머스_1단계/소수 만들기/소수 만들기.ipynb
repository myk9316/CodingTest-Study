```python
from itertools import combinations

def is_prime(num):
    if num > 1:
        for n in range(2, (num//2)+1):
            if num % n == 0:
                return False
        return True

    
def solution(nums):
    answer = 0
    a = list(combinations(nums, 3))
    for i in a:
        if is_prime(sum(i)):
            answer += 1

    return answer
```

```python
print(solution([1,2,3,4]))
print(solution([1,2,7,6,4]))
```
