# 약수의 개수와 덧셈

```python
from IPython.display import Image
Image("./문제설명.png")
```

```python
def solution(left, right):
    answer = 0
    for i in range(left, right+1):
        count = 0
        for j in range(1, i+1):
            if i% j == 0:
                count += 1
        if count % 2 == 0:
            answer +=  i
        else:
            answer -= i
                
    return answer
```

```python
print(solution(13, 17))
print(solution(24, 27))
```
