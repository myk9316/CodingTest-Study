```python
import itertools

def solution(numbers):
    answer = []
    a = itertools.combinations(numbers, 2)
    for i in a:
        answer.append(sum(i))
    answer = set(answer)
    return sorted(answer)
```

```python
print(solution([2,1,3,4,1]))
print(solution([5,0,2,7]))
```
