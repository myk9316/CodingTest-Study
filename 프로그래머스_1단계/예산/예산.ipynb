```python
def solution(d, budget):
    answer = 0
    d = sorted(d)
    for amount in d:
        if amount <= budget:
            answer += 1
            budget -= amount
        else:
            break
    return answer
```

```python
print(solution([1,3,2,5,4], 9))
print(solution([2,2,3,3], 10))
```
