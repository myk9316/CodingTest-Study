# 소수 찾기

```python
from IPython.display import Image
Image("./문제설명.png")
```

```python
def solution(n):
    num = set(range(2, n+1))
    
    for i in range(2, n+1):
        if i in num:
            num -= set(range(2*i, n+1, i))
            
    return len(num)
```

```python
print(solution(10))
print(solution(5))
```
