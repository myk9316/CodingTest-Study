```python
# Answer
def solution(s):
    answer = ''
    if len(s) % 2 == 1:
        a = len(s)//2 + 1
        answer += s[a-1]
    else:
        a = len(s)//2
        answer += s[a-1]+s[a]

    return answer
```

```python
print(solution("abcde"))
print(solution("qwer"))
```
