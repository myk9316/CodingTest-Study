```python
def solution1(a, b):
    answer = 0
    for a_num, b_num in zip(a,b):
        answer += (a_num * b_num)
    return answer
```

```python
print(solution1([1,2,3,4], [-3,-1,0,2]))
print(solution1([-1,0,1], [1,0,-1]))
```

```python
def solution2(a, b):
    return sum([x*y for x, y in zip(a,b)])
```

```python
print(solution2([1,2,3,4], [-3,-1,0,2]))
print(solution2([-1,0,1], [1,0,-1]))
```
