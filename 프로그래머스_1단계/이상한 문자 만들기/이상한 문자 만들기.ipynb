# 이상한 문자 만들기

```python
def solution(s):
    words =  s.split(" ")
    new_list = []
    for word in words:
        new_words = ""
        for i in range(len(word)):
            new_words += word[i].upper() if i%2 == 0 else word[i].lower()
        new_list.append(new_words)
    
    return " ".join(new_list)
```

```python
solution("try hello world")
```
