```python
def solution1(absolutes, signs):
    answer = 0
    for i in range(0, len(absolutes)):
        if signs[i] == True:
            answer += absolutes[i]
        else:
            answer -= absolutes[i]
            
    return answer
```

```python
print(solution1([4,7,12],[True, False, True]))
print(solution1([1,2,3],[False,False,True]))
```

```python
def solution2(absolutes, signs):
    answer = 0
    for absolute, sign in zip(absolutes, signs):
        if sign:
            answer += absolute
        else:
            answer -= absolute
            
    return answer
```

```python
print(solution2([4,7,12],[True, False, True]))
print(solution2([1,2,3],[False,False,True]))
```
