# 행렬의 덧셈

```python
from IPython.display import Image
Image("./문제설명.png")
```

```python
def solution(arr1, arr2):
    
    for i in range(len(arr1)):
        for j in range(len(arr1[0])):
            arr1[i][j] += arr2[i][j]

    return arr1
```

```python
print(solution([[1,2],[2,3]], [[3,4],[5,6]]))
print(solution([[1],[2]], [[3],[4]]))
```
