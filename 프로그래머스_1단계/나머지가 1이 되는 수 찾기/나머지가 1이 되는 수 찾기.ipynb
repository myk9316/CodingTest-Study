# 나머지가 1이 되는 수 찾기

```python
from IPython.display import Image
Image("./문제설명.png")
```

```python
def solution1(n):
    answer = []
    for i in range(1, n):
        if n % i == 1:
            answer.append(i)
    return min(answer)
```

```python
print(solution1(10))
print(solution1(12))
```

```python
def solution2(n):
    return min([x for x in range(1, n) if n%x==1])
```

```python
print(solution2(10))
print(solution2(12))
```
