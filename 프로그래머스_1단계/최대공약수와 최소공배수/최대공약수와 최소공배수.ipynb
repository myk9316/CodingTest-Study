```python
def solution1(n, m):
    answer = []
    
    smaller, bigger = min(n, m), max(n, m)
    
    gcd_candidate = [i for i in range(1, smaller+1) if smaller%i == 0][::-1]
    for j in gcd_candidate:
        if bigger % j == 0:
            answer.append(j)
            break
        else:
            continue
            
    lcm = bigger if bigger % smaller == 0 else smaller * bigger
    answer.append(lcm)

    return answer
```

```python
print(solution1(3, 12))
print(solution1(2, 5))
print(solution1(115, 12))
```

```python
from math import gcd
def solution2(n, m):
    g = gcd(n,m)
    l = (n*m)//g
    return [g,l]
```

```python
print(solution2(3, 12))
print(solution2(2, 5))
print(solution2(115, 12))
```
