```python
def solution(n):
    answer = ''
    for i in range(1, n+1):
        if i % 2 == 1:
            answer += '수'
        else:
            answer += '박'
    return answer
```

```python
solution(3)
```

```python
solution(4)
```

```python
def solution(n):
    return ('수박' * (n//2+1))[:n]
```

```python
solution(3)
```

```python
solution(4)
```
