```python
def solution(progresses, speeds):
    answer = []
    count = 0
    
    while len(progresses) > 0:
        if progresses[0] >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        
        else:
            progresses = [x+y for x,y in zip(progresses, speeds)]
            if count > 0:
                answer.append(count)
                count=0
            
    answer.append(count)       
    return answer
```

```python
print(solution([93, 30, 55], [1, 30, 5]))
print(solution([95, 90, 99, 99, 80, 99], [1, 1, 1, 1, 1, 1]))
```
