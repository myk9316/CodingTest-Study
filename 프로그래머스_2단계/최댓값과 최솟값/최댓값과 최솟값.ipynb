# 최댓값과 최솟값


```python
from IPython.display import Image
Image("./문제설명.png")
```

```python
def solution(s):
    numbers = list(map(int, s.split(' ')))

    return str(min(numbers)) + ' ' + str(max(numbers))
```

```python
print(solution("1 2 3 4"))
print(solution("-1 -2 -3 -4"))
print(solution("-1 -1"))
```
