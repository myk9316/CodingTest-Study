```python
def solution(n):
    num = ['1', '2', '4']
    answer = ''
    
    while n > 0:
        n -= 1
        answer = num[n%3] + answer
        n //= 3
        
    return answer
```

```python
print(solution(1))
print(solution(2))
print(solution(3))
print(solution(4))
```
