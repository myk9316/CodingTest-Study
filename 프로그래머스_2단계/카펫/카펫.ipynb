```python
def solution(brown, yellow):
    total = brown + yellow
    
    for weight in range(total, 2, -1):
        height = total // weight
        
        if yellow == (weight - 2) * (height - 2):
            return [weight, height]
```

```python
print(solution(10, 2))
print(solution(8, 1))
print(solution(24, 24))
```
