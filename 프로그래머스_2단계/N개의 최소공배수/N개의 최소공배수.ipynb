# N개의 최소공배수

```python
import math

def solution(arr):
    answer = arr[0]
    for n in arr:
        answer = (n * answer) // math.gcd(n, answer)
    return answer
```

```python
print(solution([2,6,8,14]))
print(solution([1,2,3]))
```
