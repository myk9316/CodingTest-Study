# 이진 변환 반복하기

```python
def solution(s):
    cnt = 0
    zero = 0
    
    while True:
        if s == '1':
            break
            
        zero += s.count('0')
        s = s.replace('0', '')
        s = bin(len(s))[2:]
        cnt += 1

    return [cnt, zero]
```

```python
print(solution("110010101001"))
print(solution("01110"))
print(solution("1111111"))
```
