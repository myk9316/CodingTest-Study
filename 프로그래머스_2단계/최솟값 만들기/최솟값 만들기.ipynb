```python
def solution(A,B):
    answer = 0
    
    A, B = sorted(A), sorted(B, reverse=True)
    
    for i in range(len(A)):
        answer += (A[i] * B[i])
    
    return answer
```

```python
print(solution([1,4,2], [5,4,4]))
print(solution([1,2], [3,4]))
```

```python
def solution(A,B):
    return sum(a*b for a,b in zip(sorted(A), sorted(B, reverse=True)))
```

```python
print(solution([1,4,2], [5,4,4]))
print(solution([1,2], [3,4]))
```
