```python
from itertools import permutations

def solution(numbers):
    number = list(numbers)
    new = []
    answer = 0
    
    for i in range(1, len(number)+1):
        new.extend(list(map(''.join, permutations(number, i))))

    new_list = list(set(list(map(int, new))))

    for a in new_list:
        if a > 1:
            for i in range(2, a):
                if a % i == 0:
                    break
            else:
                answer += 1
            
    return answer
```

```python
print(solution('17'))
print(solution('011'))
```
